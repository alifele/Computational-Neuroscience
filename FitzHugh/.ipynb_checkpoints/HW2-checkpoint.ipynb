```python
# 1)
from brian2 import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
start_scope()
num_neurons = 1
tay = 2*second

Cm =  0.001*ufarad
gl = 10e-14*siemens
El =  -65*mV
EK =  -90*mV
ENa =  50*mV
g_na = 2e-7*siemens
g_kd = 6e-6*msiemens
VT = -63*mV
```

```python
eqs = Equations('''
dv/dt = (gl*(El-v) - g_na*(m*m*m)*h*(v-ENa) - g_kd*(n*n*n*n)*(v-EK) + I)/Cm : volt
dm/dt = 0.32*(mV**-1)*4*mV/exprel((13.*mV-v+VT)/(4*mV))/ms*(1-m)-0.28*(mV**-1)*5*mV/exprel((v-VT-40.*mV)/(5*mV))/ms*m : 1
dn/dt = 0.032*(mV**-1)*5*mV/exprel((15.*mV-v+VT)/(5*mV))/ms*(1.-n)-.5*exp((10.*mV-v+VT)/(40.*mV))/ms*n : 1
dh/dt = 0.128*exp((17.*mV-v+VT)/(18.*mV))/ms*(1.-h)-4./(1+exp((40.*mV-v+VT)/(5.*mV)))/ms*h : 1
I : amp
''')
```

```python
# Threshold and refractoriness are only used for spike counting
group = NeuronGroup(num_neurons, eqs,threshold='v > -50*mV',refractory='v > -50*mV',method='exponential_euler')
group.v = El
group.I = 200*pA
M1= StateMonitor(group, 'v', record=0)
#monitor = SpikeMonitor(group)
run(100*ms)
plot(M1.t/ms, M1.v[0])
xlabel('Time (ms)')
ylabel('v');
show()
```

```python
M2= StateMonitor(group, 'm', record=0)
run(100*ms)
plot(M2.t/ms, M2.m[0])
xlabel('Time (ms)')
ylabel('m');
show()
```

```python
M3= StateMonitor(group, 'n', record=0)
run(100*ms)
plot(M3.t/ms, M3.n[0])
xlabel('Time (ms)')
ylabel('n');
show()
```

```python
M4= StateMonitor(group, 'h', record=0)
run(100*ms)
plot(M4.t/ms, M4.h[0])
xlabel('Time (ms)')
ylabel('h');
show()
```

```python
import numpy as np
import draw_utils



def F(v,w, params):
    return v-v**3-w+params[0]

def G(v,w, params):
    return 0.08*(v+0.7-0.8*w)

def f_null1(v,params):
    return (v+0.7-params[0])/(0.8*(1-v**2))

def f_null2(v,params):
    return (v+0.7)/0.8
```

```python
params=[0]
draw_utils.plot_phaseplane(F, G, f_null1, f_null2, extra_params=params,x_range=[0,3], y_range=[0,3])
```
